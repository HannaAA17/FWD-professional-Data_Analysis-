```python
import glob
import pandas as pd
```

```python
# List of dictionaries to build file by file and later convert to a DataFrame
df_list = []
#knowing that the "ebert_reviews" is the folder name
#we use glop to get the directory of all .txt files in 'ebert_reviews' folder
for ebert_review in glob.glob('ebert_reviews/*.txt'):
    #as these files in 'utf-8' format
    with open(ebert_review, encoding='utf-8') as file:
        #read the first line
        title = file.readline().rstrip()
        
        #read the next line 
        review_url = file.readline().rstrip()
        
        #read all the remaining text
        review_text = file.read().rstrip()
        
        # Append to list of dictionaries to later convert to a DataFrame
        df_list.append({'title': title,
                        'review_url': review_url,
                        'review_text': review_text})
df = pd.DataFrame(df_list, columns = ['title', 'review_url', 'review_text'])
```
